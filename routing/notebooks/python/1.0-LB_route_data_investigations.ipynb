```python
import pandas as pd
import matplotlib
import datetime

file_path = '/home/mint/Downloads/top_dest_output.csv'

df = pd.read_csv(file_path, names=['MMSI', 'Start_cell', 'End_cell', 'Start_time', 'End_time'])
df.head()
```

```python
df.describe('')
```

```python
route_groupedDF = df[["Start_cell", "End_cell", "MMSI"]].groupby(["Start_cell", "End_cell"]).agg('count')
```

```python
route_groupedDF = route_groupedDF.sort_values('MMSI', ascending=False)
dfroute_grouped = route_groupedDF.rename(columns={'MMSI':'Count'})
```

```python
dfroute_grouped
```

```python
df['Start_time'] = pd.to_datetime(df['Start_time'])
df['End_time'] = pd.to_datetime(df['End_time'])
df['End_time']-df['Start_time']
df['journey_length'] = df['End_time']-df['Start_time']
```

```python
df.sort_values('journey_length', ascending=False).head()
```

```python
df ['epoch_journey_length'] = df.journey_length.astype('timedelta64[s]')
```

```python
dflong_route = df[df.epoch_journey_length > 86400]
dflong_route.sort_values('epoch_journey_length', ascending=False)
```

```python
dflong_route
```

```python
dfstats_long_route['min'] = dflong_route[["Start_cell", "End_cell", "epoch_journey_length"]].groupby(["Start_cell", "End_cell"]).agg(min)
dfstats_long_route['max'] = dflong_route[["Start_cell", "End_cell", "epoch_journey_length"]].groupby(["Start_cell", "End_cell"]).agg(max)
dfstats_long_route['sum'] = dflong_route[["Start_cell", "End_cell", "epoch_journey_length"]].groupby(["Start_cell", "End_cell"]).agg(sum)
dfstats_long_route['count'] = dflong_route[["Start_cell", "End_cell", "epoch_journey_length"]].groupby(["Start_cell", "End_cell"]).agg('count')
dfstats_long_route['mean'] = dflong_route[["Start_cell", "End_cell", "epoch_journey_length"]].groupby(["Start_cell", "End_cell"]).agg('mean')
dfstats_long_route
```
